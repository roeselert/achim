```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf

# Download historical data from Yahoo Finance
data = yf.download("BTC-USD", start="2018-01-01", end="2023-12-24")

# Define the short and long moving average windows
short_window = 30
long_window = 40

# Calculate the short and long moving averages
data['Short_MA'] = data['Adj Close'].rolling(window=short_window, min_periods=1, center=False).mean()
data['Long_MA'] = data['Adj Close'].rolling(window=long_window, min_periods=1, center=False).mean()

# Generate the trading signals
data['Buy'] = np.where((data['Short_MA'] > data['Long_MA']) & (data['Short_MA'].shift(1) < data['Long_MA'].shift(1)), 1, 0)
data['Sell'] = np.where((data['Short_MA'] < data['Long_MA']) & (data['Short_MA'].shift(1) > data['Long_MA'].shift(1)), -1, 0)
data['Signal'] = data['Buy'] + data['Sell']

# Calculate the returns of the strategy
data['Return'] = np.log(data['Adj Close'] / data['Adj Close'].shift(1))
data['Strategy_Return'] = data['Return'] * data['Signal'].shift(1)
data['Cumulative_Return'] = np.exp(data['Return'].cumsum())
data['Cumulative_Strategy_Return'] = np.exp(data['Strategy_Return'].cumsum())

# Plot the results
plt.figure(figsize=(15,10))
plt.subplot(2,1,1)
plt.plot(data['Adj Close'], label='Price')
plt.plot(data['Short_MA'], label='Short MA')
plt.plot(data['Long_MA'], label='Long MA')
plt.scatter(data.index, data['Adj Close'].where(data['Buy'] == 1), marker='^', color='green', label='Buy')
plt.scatter(data.index, data['Adj Close'].where(data['Sell'] == -1), marker='v', color='red', label='Sell')
plt.legend()
plt.title('Bitcoin-Dollar Moving Average Crossover Strategy')
plt.subplot(2,1,2)
plt.plot(data['Cumulative_Return'], label='Buy and Hold')
plt.plot(data['Cumulative_Strategy_Return'], label='Moving Average Crossover Strategy')
plt.legend()
plt.title('Cumulative Returns')
plt.show()
```
